#### Linear Search 

```python
def linear_seach(data,target):
    for index, element in enumerate(data):
        if element == target:
            return index
    return -1
```

```python
data = [4,5.6,-1,4,67, 100,1]
```

```python
linear_seach(data, 1)
```

```python
linear_seach(data,100)
```

```python
linear_seach(data,-4)
```
